# Logical operators
In pandas, logical operators are used to combine multiple conditions when filtering data. The main ones are:

--> & for AND: both conditions must be true.

--> | for OR: at least one condition must be true.

--> ~ for NOT: reverses the condition.

```python
# & both conditions must be true.
# so we want employees where age > 35 and and salary == 100000
# ❌ Wrong:

# df[df["Department"] == "IT" & df["Salary"] > 70000]

# ✅ Correct:

df[(df['Age'] > 35) & (df['Salary'] >= 90000)]   # look bellow it returns rows both contion is True otherwise ignores

```

```python
# we want experience >= 10 and salary greater than 130000
print(df[(df['Experience_Years'] >= 10) & (df['Salary'] >= 130000)])
```

```python
# | -> similar to python OR at least one condition must be true

# So we want city is bangalore or department = IT
print(df[(df['City'] == 'Bangalore ') | (df['Department'] == 'IT')])
```

```python
# we want rating greater than 4 or experience is 10
print(df[(df['Rating'] > 4) | (df['Experience_Years'] == 10)])
```

```python
# ~ (NOT)  it flips the result if result True it flips True into False if Result False it flips into True
print(df[~(df['Department'] == 'HR')])  # look here first it finds the HR departments and return True then Not flips the result True Become False and all False become True and return all no HR records
```

```python
# isin(value1,value2,etc) : it checks values wheather the value exists in the pre defined list,set,tuple
# why we use is in instead od writing multiple or conditions we write isin()
# lets look print(df[(df['Department'] == 'IT') | (df['Department'] == 'IT')]) it looks long code and some confusion
# Thats why we use isin()
```

```python
print(df[df['Department'].isin(['IT','HR','Operations'])])  # it returns IT or HR or Operations Departments
# look we write isin([]) why we us [] because we want to check multiple values so thats why we use list means list of values
# We use list tuple set also based on our preference
```

```python
print(df[df['City'].isin(['Delhi','Pune','Ahmedabad '])][['Name','City']]) # we can also select specific columns
```

```python
# 10. between() (SQL BETWEEN) both are same 
# between() : is used to check values must be given range and The boundaries are included.like startina and ending values

print(df[df['Salary'].between(70000,95000)])
```

```python
print(df[df['Age'].between(35,45)])
```

```python
 # String Filtering (LIKE)
# Startswith : returns True for rows where the string starts with the given pattern.
# Endswith  : returns True for rows where the string ends with the given pattern.
# contains  : returns True for rows where the string contains the given pattern anywhere in the value.
```

```python
# startswith(): returns True for rows where the string starts with the given pattern.
print(df[df['Name'].str.startswith('A')])
```

```python
print(df[df['Department'].str.startswith('H')])
```

```python
# endswith(): returns True for rows where the string ends with the given pattern.
print(df[df['Name'].str.endswith('a')])
```

```python
print(df[df['Department'].str.endswith('s')])
```

```python
# contains(): returns True for rows where the string contains the given pattern anywhere in the value.
print(df[df['City'].str.contains('ra')])
```

```python
print(df[df['Name'].str.contains('ra',case = False)])  #case = false means it worsk insensitive like python is a case sensitive 

# like Name Contains RA and we search ra it not works why because python is a case sensitive so thats why we write case = False nor it works ifwe weite small or capital characters
```
