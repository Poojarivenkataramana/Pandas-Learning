```python
# What is a series ?
# series is a onediomentional so it could be one single row or single column
# Example
import pandas as pd
marks = pd.Series([85,80,90,95,99])
print(marks)
```

```python
# you cam=n also give custom indexes
marks = pd.Series([85,90,97],index = ['Student1','Student2','Student3'])
print(marks)
```

```python
# What is DataFrame ?
# DataFrame is a 2D array exist multiple rows and multiple columns
# Example :
details = pd.DataFrame({
    'Name' : ['Ramana','Demon','Jack shepord'],
    'Age' : [22,43,37],
    'Marks': [97,55,89]
    
})
print(details)
```

# Relationship between DataFrame and Series
# --> Every column in a DataFrame is actually a series

```python
# suppose 
print(details['Name'])
```

# Notice the above output 
It is a Series format why because we access specific column with a column label in a DataFrame

```python
# now we read the Data from CSV (comma separated values) from existing file
df = pd.read_csv('big_employee_data.csv')
```

```python
print(df)
```

# Selecting one column or Accessing a specific column

```python
print(df['Name'])
```

```python
print(df['Salary'])
```

# Selecting multiple columns

```python
print(df[['Name','Salary']])   # to selecting multiple columns we must use two square brackets and thats tells pandas
# it is a DataFrame not a Series 
```

```python
# lets check weather it is a series or DataFrame
print(type(df[['Name','Salary']]))
```

```python
print(type(df['Name']))
```

# Now Selecting Rows
iloc -> index location : it is used to select rows or columns using index numbers


```python
print(df.iloc[0])
# And we Check it is a Series or DataFrame
print(type(df.iloc[0]))
```

```python
# if the row 0 we make DataFrame by using [[]] double square braces
print(df.iloc[[0]])# now look output Shape also Changes
print(type(df.iloc[[0]]))
```

# selecting multiple columns by using range

```python
print(df.iloc[0:10])

print(type(df.iloc[0:10]))
```

```python
# We can specific columns aslo 
print(df.iloc[:6]['Name'])
```

```python

print(df.iloc[:]) # it takes range form starting to ending
```

# what is loc
it is used to access rows and columns using labels

```python
print(df.iloc[0:5]) #look here end index also included this is one difference
```

```python
# Selecting Specific Rows and columns
print(df.loc[:7,['Name','Salary']])
# df.loc[0:2,["Name","Salary"]]
```

```python
print(type(df.loc[0:5,'Name']))
```

```python
print(df.loc[0:5,['Name','Age','Department','Email','Salary']])
```

```python
print(df.index)
```

```python
print(df.iloc[0:6].index)
```

```python
df2 = df.set_index('Name')
print(df2.head()) # look here name becomes index
```

```python
# To remove this we use reset_index
df3 = df2.reset_index()
print(df3)
```

```python
df2 = df.set_index('Name') #Name becomes index 
print(df2.loc['Geeta Rao']) # we can access record using Speciam name that are exist in DataFrame
```
