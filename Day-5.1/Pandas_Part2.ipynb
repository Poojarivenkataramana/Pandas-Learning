### Difference between `rename()` and `replace()` in pandas

| Aspect                 | `rename()`                                              | `replace()`                                                  |
|------------------------|---------------------------------------------------------|--------------------------------------------------------------|
| Main purpose           | Changes labels (column names or index names).          | Changes data values inside the cells.                       |
| Typical use            | Renaming columns like `EmpID` → `EmployeeID`.          | Changing values like `M` → `Male`, `F` → `Female`.          |
| What it acts on        | Column names and/or index names.                       | Values in one column or the entire DataFrame.               |
| Example syntax         | `df.rename(columns={'Gender': 'Sex'})`                 | `df.replace({'M': 'Male', 'F': 'Female'})`                  |
| Scope control          | Use `columns=` or `index=` to specify what to rename.  | Use column selection or `subset` / dict mapping to target.  |
| Effect on structure    | Structure (rows/columns) stays the same, only labels.  | Structure stays same, but cell contents change.             |
| Common usage in EDA    | Make column names clearer for analysis and plotting.   | Clean inconsistent or coded values in your data.            |
