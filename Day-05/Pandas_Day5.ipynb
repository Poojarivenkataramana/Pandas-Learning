# 1️⃣ What are Missing Values?

Missing values are **empty or unavailable cells** in a dataset.

Pandas represents missing values as:

```python
NaN
```

**NaN** stands for **Not a Number**.

---

### 📌 Example

| Student_ID | Name   | Age | City    | Score |
|------------|--------|-----|---------|-------|
| 1 | Amit   | 21 | Delhi   | 85 |
| 2 | Priya  | NaN | Mumbai | 90 |
| 3 | Rahul  | 22 | NaN | 78 |
| 4 | Sneha  | 20 | Pune | NaN |
| 5 | Vikram | 23 | Chennai | 88 |

Here:
- `Age` is missing for **Priya**.
- `City` is missing for **Rahul**.
- `Score` is missing for **Sneha**.

---

## 🎯 Why are Missing Values Important?

Missing values can:
- Produce incorrect analysis.
- Affect calculations like **mean**, **sum**, and **average**.
- Reduce machine learning model performance.
- Need to be handled before data analysis.

💡 **Next:** We'll learn how to detect missing values using `isnull()`.

````python
# 2️⃣ isnull()
# Definition

# Returns True for missing values and False otherwise.

# Syntax

df.isnull()
````

````python
# Example
print(df.isnull())
````

````python
# Count Missing Values
print(df.isnull().sum())
# This is one of the most common interview questions.
````

````python
# 3️⃣ notnull()

# Opposite of isnull().

df.notnull()

# Returns

# True → value exists
# False → missing
````

````python
print(df.notnull())
````

````python
print(df.notnull().sum())  # it counts the non null values
````

````python
# 4️⃣ fillna()
# Instead of deleting missing values, we can replace them.
# so we replace null values with 18

# Step 1: fill nulls
df['age'] = df['age'].fillna(18)

# Step 2: convert to int using .astype(Datatype)
df['age'] = df['age'].astype(int)

print(df)
print("Age successfully converted Float to Integer ✅")
````

````python
# Now we fill null values in city NaN to Unknown
# Step 1 fill nulls
df['city'] = df['city'].fillna('Unknown')
print(df)
print("successfully replaced null values with Unknown ✅")
````

````python
# Now we replace Score null values with average(mean) of score
df['score'] = df['score'].fillna(df['score'].mean()) # mean is 81.75 so all nulls replaced with 81.75 in the scores
print(df)
print('Scores Null Values Successfully replaced ✅')
````

````python
# Lets look now any null present in the dataframe
print(df.isnull().sum())
print('We successfully cleaned all null Values with some Reasonable Values ✅')
````

````python
# Now we also check all non null count
print(df.notnull().sum())
````

````python
print(df.info())
````

````python
# now we rename the column headings
df.rename(columns = {
    'student_id' : 'Student_ID',
    'name' : 'Name',
    'age' : 'Age',
    'city' : 'City',
    'score': 'Scores',
    'grade' : 'Grades'
},inplace = True)
print(df)
print('All Column Headings Changes Professional Characters ✅')
````

````python
# Finally now we need to save The Cleaned Dataset we need to name the cleaned DataSet Called Student_Processed_Dataset

df.to_csv('Students_Processed_Dataset.csv', index=False)
print("Saved successfully without index ✅") # this is new Dataset Now look below

df1 = pd.read_csv('students_practice.csv')
print("This is Messy Dataset")
print(df1)

print("\n")

df2 = pd.read_csv('Students_Processed_Dataset.csv')
print("This is Cleaned Dataset ✅")
print(df2)

````

**Answer:** Because `fillna()` returns a new DataFrame unless `inplace=True` is used or the result is assigned back.
