# 📊 Sorting Data
### 🔹 Sort by One Column

> 💡 **Syntax**
>
> `df.sort_values(by="column_name", ascending=True)`
>
> 🟦 SQL

ORDER BY Salary DESC;

🐼 Pandas

df.sort_values(by="Salary", ascending=False)

````python
print(df.sort_values(by = 'Salary',ascending = False))
````

````python
print(df.sort_values(by = 'Salary',ascending = True)) # sorts lowest salary to highest
````

ascending=True
⬇
Small → Large
1000 → 2000 → 3000

ascending=False
⬇
Large → Small
3000 → 2000 → 1000



⭐ **Remember**

✔ `ascending=True` → Lowest to Highest (Default)

✔ `ascending=False` → Highest to Lowest


## 🎯 Quick Takeaway

✅ `sort_values()` sorts by column values.

✅ Default is `ascending=True`.

✅ Use `ascending=False` for descending order.

✅ Sort multiple columns using a list.

````python
# 2. Sort by Multiple Columns

# Suppose you want:

# Department ascending
# Salary descending

print(df.sort_values(by = ['Department','Salary'],
                    ascending = [True,False]))
````

> ## 🧠 Understanding Multi-Column Sorting

### 📌 What happens?

1️⃣ **First**, Pandas sorts the **Department** column in alphabetical order (A → Z).

2️⃣ **Then**, within each department, it sorts the **Salary** column from **Highest → Lowest**.

---

### 📊 Example

#### 🔹 Before Sorting

| Department | Salary |
|------------|-------:|
| IT | 50,000 |
| HR | 45,000 |
| IT | 80,000 |
| IT | 70,000 |

⬇️

#### ✅ After Sorting

| Department | Salary |
|------------|-------:|
| HR | 45,000 |
| IT | 80,000 |
| IT | 70,000 |
| IT | 50,000 |

---

💡 **Key Point**

- First key → **Department (A → Z)**
- Second key → **Salary (High → Low)**

🎯 **Pandas Code**

```python
df.sort_values(
    by=["Department", "Salary"],
    ascending=[True, False]
)
```


````python
# 3. Sort by Index

# Suppose rows become shuffled.
# syntax():

df.sort_index(ascending = True/False)
````

````python
# Sorting index highest to lowest
print(df.sort_index(ascending = False))
````

````python
# Sorting index lowest to highest
print(df.sort_index())   # here we dont need to write ascending = True because by default its ascending order (samllest to highest)
````

## ✏️ Rename Columns

> 🎯 **Goal:** Change one or more column names without changing the data.

---

### 📌 Example

**Current column name**

```text
Salary
```

⬇️ Rename to ⬇️

```text
Monthly_Salary
```

---

### 📝 Syntax

```python
df.rename(
    columns={
        "Old_Column1": "New_Column1",
        "Old_Column2": "New_Column2"
    }
)
```

---

### 💡 Example

```python
df.rename(columns={"Salary": "Monthly_Salary"})
```

---

### ⭐ Key Points

✅ Changes only the column name.

✅ Data inside the column remains the same.

✅ Returns a **new DataFrame** by default.

✅ Use `inplace=True` to modify the original DataFrame.

---

````python
# Modifing salary into monthly salary
df1 = df.rename(columns = {'Salary' : 'Monthly_Salary'})
print(df1) # this is modified one
print(df) # is a original one we didnt change original
````

````python
# Notice:

# This does not change df.

# Why?

# Because Pandas returns a new DataFrame.
print(df) # this is original
````

````python
# we want to modify original then we must use inplace = True this tells pandas to modify original
print(df.rename(
    columns = {'Salary': 'M_Salary'},inplace = True
               ))  # look output none because it return any thing it modifies original one so look below code
````

````python
df.rename(
    columns = {'Salary': 'M_Salary'},inplace = True
)
print(df) # Now original  Dataset Modified Modified
````

````python
# 5. Why doesn't Pandas change the original DataFrame?

# Most Pandas methods return a new DataFrame instead of modifying the original one.

# This helps prevent accidental data loss.
df.rename(
    columns = {'M_Salary': 'Salary'},inplace = True
)
print(df)  #now we again changed M_Salary to Salary
````

## ➕ Create a New Column

> 🎯 **Goal:** Add a new column by performing calculations on existing columns.

---

### 📌 Example

Suppose every employee receives a **10% bonus**.

**Existing Column**

| Salary |
|--------:|
| 50,000 |
| 60,000 |
| 70,000 |

⬇️ Calculate Bonus ⬇️

**New Column**

| Salary | Bonus |
|--------:|------:|
| 50,000 | 5,000 |
| 60,000 | 6,000 |
| 70,000 | 7,000 |

---

### 📝 Syntax

```python
df["New_Column"] = calculation
```

---

### 💡 Example

```python
df["Bonus"] = df["Salary"] * 0.10
```

---

### ⭐ Key Points

✅ Creates a brand-new column.

✅ Original columns remain unchanged.

✅ Calculations are performed row by row.

✅ You can use mathematical operators (`+`, `-`, `*`, `/`, `%`, `**`) and existing columns.

---

### ⚡ Output

A new **Bonus** column is added to the DataFrame.

---

### 🎯 Interview Tip

Creating a new column does **not** modify existing columns—it simply appends a new column to the DataFrame.

````python
# 6. Create a New Column

# Suppose everyone gets a bonus of 10%.

df['Bonus'] = df['Salary'] * 0.10
print(df)  # look at the dataframe new column appeared called Bonus
````

````python
# 7. Create Another Column

# Net Salary
df['Net_Salary'] = df['Salary'] + df['Bonus']
print(df.head()) #look another column added called Net_Salary
````

````python
# SO no we update existing Salary COlumn by 0.5%
df['Salary'] = df['Salary'] * 1.05
print(df)
````

````python
df['Salary'] = df['Salary'] * 1.05
df['Salary'] = df['Salary'] + (df['Salary'] * 0.5) 
# Both are gives same result as above but we prefer  df['Salary'] = df['Salary'] * 1.05 because its short one
````

## 🗑️ Drop a Column

> 🎯 **Goal:** Remove one or more columns from a DataFrame.

---

### 📌 Example

Suppose the **Bonus** column is no longer required.

**Before**

| Name | Salary | Bonus |
|------|--------:|------:|
| Ram | 50,000 | 5,000 |
| Ravi | 60,000 | 6,000 |

⬇️ Drop the `Bonus` column ⬇️

**After**

| Name | Salary |
|------|--------:|
| Ram | 50,000 |
| Ravi | 60,000 |

---

### 📝 Syntax

```python
df.drop(columns=["Column_Name"])
```

Drop multiple columns:

```python
df.drop(columns=["Column1", "Column2"])
```

---

### 💡 Example

```python
df.drop(columns=["Bonus"])
```

---

### ⭐ Key Points

✅ Removes the specified column(s).

✅ By default, returns a **new DataFrame**.

✅ The original DataFrame is **not modified**.

✅ To permanently remove the column:

```python
df.drop(columns=["Bonus"], inplace=True)
```

---

### ⚠️ Interview Tip

- `drop()` → Returns a new DataFrame.
- `drop(..., inplace=True)` → Modifies the original DataFrame.

````python
# Suppose Bonus is no longer needed.

print(df.drop(columns=["Bonus"])) #it returns a new DataFrame but we want to remove original

print('Original DataFrame'.center(70,'-'))
print(df) # look original not changed
print('-' * 70)

# This returns a new DataFrame.

# To make it permanent:
````

````python
# So if we Drop a Original DataFrame Column use inplace function
df.drop(columns = ['Bonus'],inplace = True)
print(df)
````

````python
# We also remove Net_Salary
df.drop(columns = ['Net_Salary'],inplace = True)
print(df)
````

````python
# we can also delete multiple columns 
# like rating and join_Data
df2 = df.drop(columns = ['Rating','JoinDate'])
print(df2) #look below
````

# ⚡ Rapid Fire Interview Questions & Answers

**1. Which function is used to sort rows in Pandas?**  
✅ `sort_values()`

**2. Which function is used to sort the index?**  
✅ `sort_index()`

**3. What is the default value of `ascending`?**  
✅ `True`

**4. How do you sort in descending order?**  
✅ `ascending=False`

**5. Can `sort_values()` sort multiple columns?**  
✅ Yes, by passing a list to the `by` parameter.

**6. Which parameter specifies the column name while sorting?**  
✅ `by`

**7. What does `ascending=[True, False]` mean?**  
✅ First column ascending, second column descending.

**8. Which function is used to rename columns?**  
✅ `rename()`

**9. Does `rename()` modify the original DataFrame by default?**  
✅ No. It returns a new DataFrame.

**10. How do you modify the original DataFrame while renaming?**  
✅ Use `inplace=True`.

**11. Which approach is preferred: `inplace=True` or assignment?**  
✅ Assignment (`df = ...`) is generally preferred.

**12. How do you create a new column?**  
✅ `df["New_Column"] = value`

**13. Can a new column be created from existing columns?**  
✅ Yes.

**14. How do you update an existing column?**  
✅ `df["Salary"] = df["Salary"] * 1.05`

**15. What is the difference between creating and updating a column?**  
✅ Creating adds a new column; updating changes an existing one.

**16. Which function removes a column?**  
✅ `drop()`

**17. Does `drop()` modify the original DataFrame by default?**  
✅ No. It returns a new DataFrame.

**18. How do you permanently remove a column?**  
✅ `df.drop(columns=["Bonus"], inplace=True)`

**19. Can multiple columns be dropped at once?**  
✅ Yes.

**20. What happens if you drop a column that doesn't exist?**  
✅ A `KeyError` is raised (unless `errors="ignore"` is used).

**21. Which function sorts by row labels (index)?**  
✅ `sort_index()`

**22. What is the default axis in `drop()`?**  
✅ `axis=0` (rows), but `columns=` is preferred for dropping columns.

**23. Can `rename()` rename multiple columns at once?**  
✅ Yes.

**24. Does creating a new column delete existing columns?**  
✅ No.

**25. Which method is similar to SQL `ORDER BY`?**  
✅ `sort_values()`

**26. Which Pandas method is similar to SQL `AS`?**  
✅ `rename()`

**27. Which method is similar to SQL `ALTER TABLE DROP COLUMN`?**  
✅ `drop(columns=...)`

**28. How do you increase every salary by 5%?**  
✅ `df["Salary"] = df["Salary"] * 1.05`

**29. How do you calculate a 10% bonus column?**  
✅ `df["Bonus"] = df["Salary"] * 0.10`

**30. What does `inplace=True` mean?**  
✅ Modify the original DataFrame instead of returning a new one.
